### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja = float(sum(loja['Preço']))
faturamento_loja2 = float(sum(loja2['Preço']))
faturamento_loja3 = float(sum(loja3['Preço']))
faturamento_loja4 = float(sum(loja4['Preço']))

print(f"Faturamento Loja 1: R$ {faturamento_loja:,.2f}")
print(f"Faturamento Loja 2: R$ {faturamento_loja2:,.2f}")
print(f"Faturamento Loja 3: R$ {faturamento_loja3:,.2f}")
print(f"Faturamento Loja 4: R$ {faturamento_loja4:,.2f}")
```

```python
faturamentos = {
    'Loja 1': faturamento_loja,
    'Loja 2': faturamento_loja2,
    'Loja 3': faturamento_loja3,
    'Loja 4': faturamento_loja4 }

faturamentos_ordenados = dict(sorted(faturamentos.items(), key=lambda item: item[1]))

x = np.array(list(faturamentos_ordenados.keys()))
y = np.array([float(valor) for valor in faturamentos_ordenados.values()])

plt.bar(x, y, color='skyblue')
plt.xlabel('Lojas')
plt.ylabel('Faturamento')
plt.ylim(1300000.0000000000,1600000.0000000000)
plt.text(x[0], y[0], f'R$ {y[0]:,.2f}', ha='center', va='bottom')
plt.text(x[1], y[1], f'R$ {y[1]:,.2f}', ha='center', va='bottom')
plt.text(x[2], y[2], f'R$ {y[2]:,.2f}', ha='center', va='bottom')
plt.text(x[3], y[3], f'R$ {y[3]:,.2f}', ha='center', va='bottom')
plt.title('Faturamento por Loja')
plt.show()
```

# 2. Vendas por Categoria


```python
df = pd.read_csv(url)
contagem_loja = df ["Categoria do Produto"].value_counts()

df = pd.read_csv(url2)
contagem_loja2 = df ["Categoria do Produto"].value_counts()

df = pd.read_csv(url3)
contagem_loja3 = df ["Categoria do Produto"].value_counts()

df = pd.read_csv(url4)
contagem_loja4 = df ["Categoria do Produto"].value_counts()
```

```python
dict_loja1 = contagem_loja.to_dict()
dict_loja2 = contagem_loja2.to_dict()
dict_loja3 = contagem_loja3.to_dict()
dict_loja4 = contagem_loja4.to_dict()

#Grafico Loja1

plt.figure(figsize=(4, 4))
plt.pie(dict_loja1.values(), labels=dict_loja1.keys(), autopct='%1.1f%%',)


plt.title('Distribuição de Vendas por Categoria - Loja 1')
plt.axis('equal')
plt.show()


#Grafico Loja2
plt.figure(figsize=(4, 4))
plt.pie(dict_loja2.values(), labels=dict_loja2.keys(), autopct='%1.1f%%',)


plt.title('Distribuição de Vendas por Categoria - Loja 2')
plt.axis('equal')
plt.show()

#Grafico Loja3
plt.figure(figsize=(4, 4))
plt.pie(dict_loja3.values(), labels=dict_loja3.keys(), autopct='%1.1f%%',)


plt.title('Distribuição de Vendas por Categoria - Loja 3')
plt.axis('equal')
plt.show()


#Grafico Loja4
plt.figure(figsize=(4, 4))
plt.pie(dict_loja4.values(), labels=dict_loja4.keys(), autopct='%1.1f%%',)


plt.title('Distribuição de Vendas por Categoria - Loja 4')
plt.axis('equal')
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
media_avaliacao_loja = float(sum(loja['Avaliação da compra']) / len(loja['Avaliação da compra']))
media_avaliacao_loja2 = float(sum(loja2['Avaliação da compra']) / len(loja2['Avaliação da compra']))
media_avaliacao_loja3 = float(sum(loja3['Avaliação da compra']) / len(loja3['Avaliação da compra']))
media_avaliacao_loja4 = float(sum(loja4['Avaliação da compra']) / len(loja4['Avaliação da compra']))
```

```python
avaliacoes = {
    'Loja 1': media_avaliacao_loja,
    'Loja 2': media_avaliacao_loja2,
    'Loja 3': media_avaliacao_loja3,
    'Loja 4': media_avaliacao_loja4 }

avaliacoes_ordenados = dict(sorted(avaliacoes.items(), key=lambda item: item[1]))

x = np.array(list(avaliacoes.keys()))
y = np.array([float(valor) for valor in avaliacoes_ordenados.values()])

plt.bar(x, y, color='green')
plt.xlabel('Lojas')
plt.ylabel('Médias')
plt.ylim(3.8,4.10)
plt.text(x[0], y[0], f'{y[0]:,.2f}', ha='center', va='bottom')
plt.text(x[1], y[1], f'{y[1]:,.2f}', ha='center', va='bottom')
plt.text(x[2], y[2], f'{y[2]:,.2f}', ha='center', va='bottom')
plt.text(x[3], y[3], f'{y[3]:,.2f}', ha='center', va='bottom')
plt.title('Média das Avaliações por Loja')
plt.show()
```
